```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
#2nd fiscal quarter
df = pd.read_csv("GOOG_DS1.csv")
df.Date = df.Date.replace({'-':''}, regex=True)
df.Average_Price = (df.High + df.Low) / 2
x = np.array(df.Date).reshape((-1, 1))
y = np.array(df.Average_Price)
linreg = LinearRegression().fit(x, y)
#r2 value
r_squared = linreg.score(x, y)
print(r_squared)
#3rd fiscal quarter
df2 = pd.read_csv("GOOG_DS2.csv")
df2.Date = df2.Date.replace({'-':''}, regex=True)
df2["Average_Price"] = (df2.High + df2.Low) / 2
df2=df2.reset_index()
x = np.array(df2.Date).reshape((-1, 1))
y = np.array(df2.Average_Price)
y_pred = linreg.predict(x)
list = []
for i in range(len(y_pred)) :
    list.append(i)
plt.plot(list, y_pred)
plt.title("Predicted Google Stock Price Over Time")
plt.xlabel('Days after 6/29/22')
plt.ylabel('y_pred')
print(f"predicted response:\n{y_pred}")
plt.show()
```
